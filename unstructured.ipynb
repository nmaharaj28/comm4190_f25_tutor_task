Basic Vowels:
- A = "ah" (like in "father")

Consonants are mostly similar to English, but here are some important differences:
- C is always hard, like "k" (never soft like "s")
- G is always hard, like in "go" (never soft like in "gentle")
- V is pronounced like our "w"
- J is pronounced like "y" in "yes"

### USER
What are diphthongs?

### ASSISTANT
A diphthong is when two vowels combine to make a single sound. Think of how in English "ou" makes an "ow" sound in "house" - that's a diphthong!

In Classical Latin, there are three main diphthongs:

1. AE = "ai" (like in "aisle")
Example: CAESAR = "KAI-sar"

2. AU = "ow" (like in "how")
Example: AUDIO = "OW-dee-oh"

3. OE = "oi" (like in "oil")
Example: POENA = "POI-nah"

All other vowel combinations are pronounced separately. For example:
- EA in DEA (goddess) = "DEH-ah" (not a diphthong)
- EO in DEORUM (of the gods) = "deh-OH-room" (not a diphthong)
